```python
import pandas as pd
import numpy as np

movies=pd.read_csv('movies.csv')
actors=pd.read_csv('actors.csv')
countries=pd.read_csv('countries.csv')
crew=pd.read_csv('crew.csv')
genre=pd.read_csv('genres.csv')
languages=pd.read_csv('languages.csv')
themes=pd.read_csv('themes.csv')
```

```python
movies
```

```python
print(actors.shape)
actors.head()
```

```python
actors=actors.groupby('id').agg({'name':list,'role':list}).reset_index()
```

```python
actors.head()
```

```python
actors.shape
```

```python
countries=countries.groupby('id').agg({'country':list}).reset_index()
```

```python
print(countries.shape)
countries.head()
```

```python
crew.head()
```

```python
crew=crew.groupby('id').agg({'role':list,'name':list}).reset_index()
```

```python
crew.head()
```

```python
genre.head()
```

```python
genre=genre.groupby('id').agg({'genre':list}).reset_index()
```

```python
genre.head()
```

```python
languages.drop('type',axis=1,inplace=True)
```

```python
languages=languages.groupby('id').agg({'language':list}).reset_index()
```

```python
movies=movies.merge(crew,on='id',how='inner')
```

```python
movies.rename(columns={'role_x':'actors_role','name_x':'name','name_y':'crew_name','role_y':'crew'},inplace=True)
```
